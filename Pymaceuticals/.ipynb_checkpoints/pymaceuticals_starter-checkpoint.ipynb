```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as sts

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_data=pd.merge(study_results, mouse_metadata, on="Mouse ID", how="left")
# Display the data table for preview
combined_data
```

```python
# Checking the number of mice.
len(combined_data["Mouse ID"].unique())
```

```python
#Getting the duplicate mice by ID number that bshow up for Mouse ID and Timepoint
dupes=combined_data.loc[combined_data.duplicated(subset=["Mouse ID", "Timepoint"]), "Mouse ID"].unique()
print(dupes)
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
combined_data.loc[combined_data.loc[:, "Mouse ID"]=="g989",:]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
filter_df=combined_data.loc[:, "Mouse ID"].isin(dupes)==False
clean=combined_data.loc[filter_df,:]
clean
```

```python
# Checking the number of mice in the clean DataFrame.
len(clean["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

mean=clean[["Drug Regimen", "Tumor Volume (mm3)"]].groupby("Drug Regimen").mean()["Tumor Volume (mm3)"]
median=clean[["Drug Regimen", "Tumor Volume (mm3)"]].groupby("Drug Regimen").median()["Tumor Volume (mm3)"]
variance=clean[["Drug Regimen", "Tumor Volume (mm3)"]].groupby("Drug Regimen").var()["Tumor Volume (mm3)"]
std=clean[["Drug Regimen", "Tumor Volume (mm3)"]].groupby("Drug Regimen").std()["Tumor Volume (mm3)"]
sem=clean[["Drug Regimen", "Tumor Volume (mm3)"]].groupby("Drug Regimen").sem()["Tumor Volume (mm3)"]
# This method is the most straighforward, creating multiple series and putting them all together at the end.
Stat_table=pd.DataFrame({
    "Mean":mean,
    "Median":median,
    "Std":std,
    "Variance":variance,
    "SEM":sem,
})

Stat_table
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
statistics_dict={
    "Mean":mean,
    "Median":median,
    "Variance":variance,
    "Standard_Deviation":std,
    "Standard_Error_of_Mean":sem
}
summary_statistics=pd.DataFrame(statistics_dict)
summary_statistics
# This method produces everything in a single groupby function
summary_statistics=clean.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":["mean","median","var","std","sem"]})
summary_statistics
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
regimen=["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
# Start by getting the last (greatest) timepoint for each mouse
max_time=clean.loc[:,["Mouse ID", "Timepoint"]].groupby("Mouse ID").max()

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
max_time_study=pd.merge(max_time, clean, on=["Mouse ID", "Timepoint"], how="left")
max_time_study
```

```python
# Put treatments into a list for for loop (and later for plot labels)
regimen=["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_list=[]

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for each_drug in regimen:
    filter_df=max_time_study.loc[:, "Drug Regimen"]==each_drug
    four_quantiles=[.25, .5, .75]
    one_drug_tumor_vol=max_time_study.loc[filter_df,"Tumor Volume (mm3)"]
    tumor_vol_list.append(one_drug_tumor_vol)
    quartiles=one_drug_tumor_vol.quantile(four_quantiles)
    lowerq=quartiles.loc[.25]
    upperq=quartiles.loc[.75]
    iqr=upperq-lowerq
    lower_bound=lowerq-(1.5*iqr)
    upper_bound=upperq+(1.5*iqr)

    # Locate the rows which contain mice on each drug and get the tumor volumes
    # add subset
    # Determine outliers using upper and lower bounds
    filter_df=(one_drug_tumor_vol>upper_bound) | (one_drug_tumor_vol<lower_bound)
    outlier_count=len(one_drug_tumor_vol.loc[filter_df])
    print(f"There are {outlier_count} outliers for {each_drug}")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
plt.boxplot(tumor_vol_list, labels=regimen)
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
filter_df=clean.loc[:, "Drug Regimen"]=="Capomulin"
capomulin=clean.loc[filter_df,:]
filter_df=capomulin.loc[:,"Mouse ID"]=="b128"
one_mouse=capomulin.loc[filter_df,:]
one_mouse.plot("Timepoint", "Tumor Volume (mm3)")
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Capomulin Treatment for b128")
plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
capomulin_avg=capomulin[["Mouse ID", "Tumor Volume (mm3)", "Weight (g)"]].groupby("Mouse ID").mean()
capomulin_avg.plot.scatter("Weight (g)", "Tumor Volume (mm3)")
plt.title("Average Tumor Volume (mm3) vs. Weight (g) per Mouse")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
corr=round(sts.pearsonr(capomulin_avg["Weight (g)"], capomulin_avg["Tumor Volume (mm3)"])[0],2)
print(f"Correlation is {corr}.")
```

```python
#y=mx+b
from scipy.stats import linregress
x_value=capomulin_avg["Weight (g)"]
y_value=capomulin_avg["Tumor Volume (mm3)"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_value, y_value)
regress_value=x_value*slope+intercept
line_eq="y="+str(round(slope,2))+ "x+" +str(round(intercept,2))
plt.scatter(x_value, y_value, marker="o", color="blue")
plt.plot(x_value, regress_value, 'r-')
plt.annotate(line_eq, (15,44), fontsize=15, color="red")
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Average Tumor Volume (mm3) vs. Weight (g) per Mouse")
print(f"The r-squared value is {rvalue**2}")
plt.show()
```
